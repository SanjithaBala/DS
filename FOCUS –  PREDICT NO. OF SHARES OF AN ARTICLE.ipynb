```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
article = pd.read_csv("Datasets/OnlineNewsPopularity.csv")
print(article)
```

```python
Finalcolumns = article[:]
Finalcolumns.drop(['url'], axis=1, inplace=True)
Finalcolumns
```

```python
x=Finalcolumns.iloc[:,:-1].values
x
```

```python
y=Finalcolumns.iloc[:,1].values
y
```

```python
#Spliting the dataset into training and testing 
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=10)
```

```python
#Fit the model
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(x_train,y_train)
```

```python
#Predication of test and train
x_pred=regressor.predict(x_test)
print(x_pred)
```

```python
pd.DataFrame({'Shares of an article':x_pred,"Known output::":y_test})
```

```python
plt.figure(figsize=(12, 6))
plt.scatter(y_test, x_pred, alpha=0.3, color='blue')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], color='red', linestyle='--')
plt.title('Actual vs Predicted Shares')
plt.xlabel('Actual Shares')
plt.ylabel('Predicted Shares')
plt.show()
```
